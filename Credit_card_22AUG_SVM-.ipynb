```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(), cmap='coolwarm' )
```

```python
plt.figure(figsize=(30,25),facecolor='yellow',edgecolor='green',dpi=100)
sns.heatmap(df.corr(), cmap= 'coolwarm', annot= True)
```

```python
y = df['Class']
y.head()
print(y.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
scaled = ss.fit_transform(df.drop('Class', axis=1))
```

```python
scaled
```

```python
dff=pd.DataFrame(scaled,columns=df.columns[:-1])
dff.head()
```

```python
#Dimensionality Reduction
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(dff)
```

```python
x_pca=pca.transform(dff)
x_pca
```

```python
x_pca.shape
```

```python
plt.scatter(x_pca[y==0,0], x_pca[y==0,1],color='yellow')
plt.scatter(x_pca[y==1,0], x_pca[y==1,1], color='red')
```

```python
plt.scatter(x_pca[y==1,0], x_pca[y==1,1],color='red')
```

```python
plt.scatter(x_pca[y==0,0], x_pca[y==0,1],color='yellow')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
classifier =SVC()
```

```python
classifier.fit(X_train,y_train)
```

```python
ypred=classifier.predict(X_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
cm=confusion_matrix(ypred,y_test)
cm
```

```python
classifier.score(X_test,y_test)
```

```python
accuracy_score(ypred,y_test)
```

```python
classifier=SVC(C=10)
```

```python
classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
```

```python
classifier = SVC(gamma=20)
classifier.fit(X_train,y_train)
print(classifier.score(X_test,y_test))
plt.scatter(x_pca[y==0, 0], x_pca[y==0, 1],x_pca[y==0, 2] ,c='red')
plt.scatter(x_pca[y==1, 0], x_pca[y==1, 1], x_pca[y==1, 2],c='yellow')
```

```python
classifier = SVC(kernel='linear')#linear', 'poly', 'rbf', 'sigmoid', 'precomputed' or a callable.)

classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
plt.scatter(x_pca[y==0, 0], x_pca[y==0, 1],x_pca[y==0, 2] ,c='red')
plt.scatter(x_pca[y==1, 0], x_pca[y==1, 1],x_pca[y==1, 2], c='yellow')
```

```python
classifier = SVC(kernel='rbf',C=5.0)#linear', 'poly', 'rbf', 'sigmoid', 'precomputed' or a callable.)

classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
plt.scatter(x_pca[y==0, 0], x_pca[y==0, 1],x_pca[y==0, 2] ,c='red')
plt.scatter(x_pca[y==1, 0], x_pca[y==1, 1],x_pca[y==1, 2], c='yellow')
```

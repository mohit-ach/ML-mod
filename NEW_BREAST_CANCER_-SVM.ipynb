```python
df.drop(['worst fractal dimension','mean fractal dimension','fractal dimension error','symmetry error',
        'compactness error', 'concavity error','texture error','smoothness error'],axis=1,inplace=True)
```

```python
y= df['target']
```

```python
y.head()
```

```python
df.columns
```

```python
df.drop('target',axis=1 , inplace=True)
```

```python
df.head()
```

```python
# DIMENSIONILAITY REDUCTION
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_pca=pca.fit_transform(df)
```

```python
x_pca
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
scaled=ss.fit_transform(x_pca)
```

```python
scaled
```

```python
dff = pd.DataFrame(scaled)
```

```python
dff
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(dff,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC()
```

```python
classifier.fit(X_train,y_train)
```

```python
ypred=classifier.predict(X_test)
```

```python
ypred
```

```python
y_test
```

```python
plt.scatter(ypred,y_test)
```

```python
from sklearn.metrics import classification_report ,accuracy_score,confusion_matrix
```

```python
print(classification_report(ypred,y_test))
```

```python
print(accuracy_score(ypred,y_test))
```

```python
print(confusion_matrix(ypred,y_test))
```

```python
classifier = SVC(C=10)
```

```python
classifier.fit(X_train,y_train)
```

```python
y1pred=classifier.predict(X_test)
```

```python
y1pred
```

```python
classifier = SVC(gamma=20)
classifier.fit(X_train,y_train)
classifier.predict(X_test)
print(classifier.score(X_test,y_test))
plt.scatter(x_pca[y==0, 0], x_pca[y==0, 1])
plt.scatter(x_pca[y==1, 0], x_pca[y==1, 1], c='yellow')
```

```python
classifier = SVC(kernel='linear')#linear', 'poly', 'rbf', 'sigmoid', 'precomputed' or a callable.)

classifier.fit(X_train,y_train)
classifier.predict(X_test)
print(classifier.score(X_test,y_test))
plt.scatter(x_pca[y==0, 0], x_pca[y==0, 1])
plt.scatter(x_pca[y==1, 0], x_pca[y==1, 1], c='yellow')
```

```python
#PCA need not apply standard scaling
```

```python
X=df.drop(['target','names'], axis=1)
```

```python
X.head()
```

```python
y = df['target']
```

```python
y.head()
```

```python
# Now we have X and y
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_pca = pca.fit(X)
```

```python
x_pca=pca.transform(X)
```

```python
x_pca
```

```python
x_pca.shape
```

```python
y.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[y==0,0],x_pca[y==0,1], color='red')
plt.scatter(x_pca[y==1,0],x_pca[y==1,1],color='green')
plt.scatter(x_pca[y==2,0],x_pca[y==2,1],color='blue')
```

```python
# now we have x_pca and y
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_pca,y,test_size = 0.3)
```

```python
len(X_train),len(X_test),len(y_train),len(y_test),
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
ypred = classifier.predict(X_test)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ypred,y_test)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn import tree
plt.figure(figsize=(25,20))
tree.plot_tree(classifier, filled = True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(n_estimators=80)
rfc.fit(X_train , y_train)
```

```python
rfc_pred=rfc.predict(X_test)
rfc_pred
```

```python
rfc.score(X_train , y_train)
```

```python
rfc.score(X_test , y_test)
```

```python
rfc.estimators_
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,rfc_pred)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
plt.figure(figsize=(12,8), dpi=200)
for i in range(len(rfc.estimators_)):
    tree.plot_tree(rfc.estimators_[i], filled=True )
    
```

```python
ss= StandardScaler()
```

```python
scaled = ss.fit_transform(X)
```

```python
dff= pd.DataFrame(scaled,columns=df.columns[:-1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(dff,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
ypred=knn.predict(X_test)
```

```python
ypred
```

```python
y_test
```

```python
np.mean(y_test==ypred)
```

```python
np.mean(y_test!=ypred)
```

```python
error_rate = []
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    error_rate.append(np.mean(y_pred!=y_test))
    print(error_rate)
print(error_rate)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,20),error_rate, ls='dashed' ,lw=3,marker='o',color='red' , markerfacecolor='yellow',
         markersize=10,markeredgecolor='blue',markeredgewidth=3)
plt.title('K-VALUE')
plt.xlabel('RANGE(1,20)')
plt.ylabel('ERROR RATE')
```

```python
knn=KNeighborsClassifier(n_neighbors=14)
knn.fit(X_train,y_train)
```

```python
y1pred=knn.predict(X_test)
```
